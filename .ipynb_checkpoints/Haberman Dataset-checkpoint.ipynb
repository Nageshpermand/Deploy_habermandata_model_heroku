# Haberman's Survival Data set

### Description: 
       The Dataset contains cases from a study that was cunducted between 1958 and 1970 at the University of Chicago's Brillient Hospital on the survival of patients who had undergone surgery for breast cancer.

### Attribute information:

1. Age of Patient at time of operation (numerical).
2. Patient's year of operation(year-1990, numerical).
3. Number of positive axilary nodes detected(numerical): https://www.medicalnewstoday.com/articles/319713.php
    * lymph node  in the area of armpit to wich cancer has to spread in the human body  that nodes are positive     
       axilary nodes
          
         


**4.Survival status(class label(dependent variable) attributes)**:
     * 1 = the patient survived 5 years or longer.
     * 2 = the patient died within 5 years
    
some more information at :https://archive.ics.uci.edu/ml/datasets/Haberman%27s+Survival

## 2. CDF (Commulative Distrubution Function):

```python
#creating cure_df sub dataframe for group all features with respect to surv_status is 'cure'.
#creating fail_df sub dataframe for gruop all features with respect to surv_status is 'fail'.
cure_df = df.loc[df.surv_status == 'cure']
fail_df = df.loc[df.surv_status == 'fail']
```

**CDF is more significance for easily visual of what percentage of surv_status have axi nodes.**


```python
# plot the cdf of cure_df dataframe's axi_nodes Feature.
#construct a CDF

counts, bin_edges = np.histogram(cure_df['axi_nodes'], bins=10, density= True)

#counts datapoints in each bin edges width.
print('counts:',counts)

#bin_edges are formed by 1 st value is first value of axis_nodes next value increment with width space, next value so on so forth.
print('bin edges:',bin_edges)
pdf = counts/sum(counts)
print(pdf)



#first we need to creat pdf,  to get CDF need to cummulative sum of PDF.
#CDF commulative distributive function prcentage values on y axis. 
#CDF is equal to integration of the PDF values.
cdf = np.cumsum(pdf)


#plot pdf's bin_edges values form 1 to last value,
#plot pdf's bin_edges values from 1 to last values, 
#two plots on one figure
plt.plot(bin_edges[1:],pdf, label='PDF of (axi_nodes of survived patients' )
plt.plot(bin_edges[1:],cdf, label='CDF of(axi_nodes of survived patients')

print("*"*80)


#plot the cdf of fail_df dataframe's axi_nodes Feature.
counts, bin_edges = np.histogram(fail_df['axi_nodes'], bins = 10, density= True)

print('counts:', counts)
print('bin_edges:',bin_edges)

pdf = counts/sum(counts)
print("pdf:",pdf)

cdf = np.cumsum(pdf)
print('cdf',cdf)

plt.plot(bin_edges[1:],pdf, label='PDF of(axi_nodes of died patients)')
plt.plot(bin_edges[1:],cdf,  label='CDF of (axi_nodes of died patients)')


plt.xlabel('axi_nodes',size=13)
plt.title('PDF and CDF of axi_nodes feature',size=16)
plt.ylabel('percentage of patients',size=12)
plt.legend(loc='centre left', bbox_to_anchor=[0.5, 0.5])
plt.show()




```

### Observation:
   * blue line is pdf of cure_df ['axi_nodes'] and orange line is cdf.
   * green line is pdf of fail_df['axi_nodes'] and red line is cdf.
   
        * if axi_nodes<24: then patients survived 5 years or longer , 90% of survieved patients have axi_nodes less then 24.
     



```python
#perform CDF on anathor Feature age
# dataframe :cure_df, feature: age
counts, bin_edges = np.histogram(cure_df['age'], bins = 10, density=True)

print('counts',counts)
print('bin_edges',bin_edges)

pdf = counts/sum(counts)

print('pdf',pdf)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf,   label="PDF of(age of survived patients)" )
plt.plot(bin_edges[1:], cdf,    label="PDF of (age of survived patients)")

print("*"*80)

#dataframe = fail_df, feature:age

counts, bin_edges = np.histogram(fail_df['age'], bins = 10, density=True)
pdf  = counts/sum(counts)
print("counts:",counts)
print("bin_edges:",bin_edges)
print("pdf:",pdf)
cdf = np.cumsum(pdf)
print("cdf:",cdf)
plt.plot(bin_edges[1:], pdf, label="PDF of (age of diedpatients)")
plt.plot(bin_edges[1:], cdf, label="CDF of (age of died patients")
plt.xlabel('age',size=13)
plt.ylabel('percentage of patients')
plt.title('PDF and CDF of feature age',size=16)
plt.legend(loc='centre left', bbox_to_anchor=(1,0.5))
plt.show()
```

#### Observation:
     * cure_df of 1st  class is ( patient survived more then 5 years or longer).
     * fail_df of 2nd  class is (patient died within 5 years).
     
   * patients who have less then 37 yrs age they are survived more then 5 years or longer.
   * here 10% of (if age less then 37) patients survived.
     
       .

```python
#CDF on anathor feature operation year.
#dataframe: cure_df, feature: oprtn_yr

counts, bin_edges = np.histogram(cure_df.oprtn_yr, bins= 10, density = True)

pdf = counts/sum(counts)

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf, label="PDF of(operation year of survived patients)")
plt.plot(bin_edges[1:],cdf,  label="CDF of(operation year of survived patients)")

#dataframe: fail_df, feature:oprt_yr

counts, bin_edges = np.histogram(fail_df.oprtn_yr, bins = 10, density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf, label="PDF of (operation year of died patients)")
plt.plot(bin_edges[1:],cdf,  label="CDF of (operation year of died patients)")

plt.xlabel('oprt_yr',size=13)
plt.ylabel('percentage of  patients')
plt.title('PDF and CDF of operation year feature',size=14)
plt.legend(loc='center left', bbox_to_anchor=[1, 0.5])
plt.show()
```

#### Observation:
      1. patients who got operation in  year between 1961-1965 they are survived patients(cure).
      2. here 70% of patients survived in year between 1961-1965.
